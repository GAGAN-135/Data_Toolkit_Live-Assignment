```python
#Q17-Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
```

```python
x = np.random.randn(100)
y = np.random.randn(100)
def determine_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Quadrant I'
    elif x < 0 and y > 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'
```

```python
df = pd.DataFrame({'x': x, 'y': y})
df['Quadrant'] = df.apply(lambda row: determine_quadrant(row['x'], row['y']), axis=1)
```

```python
sns.scatterplot(df, x='x',y='y', hue='Quadrant')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrant')
plt.show()
```

```python
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']
values = np.random.randint(1, 100, size=len(categories))
df = pd.DataFrame({'categories': categories, 'values': values})
source = ColumnDataSource(df)
```

```python
#Q21-Usin Plotly, create an interactive pie chart of randomly enerated data, add labels and percentaes, set the title as 'Interactive Pie Chart'.
```

```python
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values = np.random.randint(1, 100, size=len(categories))
total = sum(values)
percentages = [(value / total) * 100 for value in values]
data = {'Categories': categories, 'Values': values, 'Percentages': percentages}
df = pd.DataFrame(data)
```

```python
fig = px.pie(df, names='Categories', values='Values',
             title='Interactive Pie Chart',
             labels={'Categories': 'Category', 'Values': 'Value'},
             hover_data=['Percentages'])
fig.show()
```
